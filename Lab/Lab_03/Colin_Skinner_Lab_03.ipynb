### Imports

```python
import string
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice
from collections import Counter
```

### First, read in the text:

```python
shakes = open("romeoandjuliet.txt").read()  
```

### Next, split up the text into individual words:

```python
text = shakes.split()
```

### To get accurate word frequencies, we need to make all characters the same case and ignore punctuation:

```python
text = [x.upper() for x in text]

translator = str.maketrans('', '', string.punctuation)
raw_text = []
for x in text:
    raw_text = raw_text + [x.translate(translator)]
```

### Now looking at the first 100 words in our modified text, all characters are capitalized and there is no punctuation:

```python
%pprint
raw_text[:100]
```

### Put the words into dictionary as keys, with their values being their frequencies in the text:

```python
word_count = {}

for i in raw_text:
    word_count[i] = raw_text.count(i)
    
    
```

### Looking at 100 items in the dictionary, we see some words followed by their frequencies:

```python

def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))

take(100, word_count.items())
```

### There are 3764  unique words in Romeo and Juliet

```python
frequencies = list(word_count.values())
words = list(word_count.keys())

print(len(words))
```

### Bar plot of frequency for all 3764 words (click to zoom in):

```python
%matplotlib inline

plt.figure(figsize =(750,8))
plt.xticks(rotation = 90)
plt.ylabel("Word Frequency")
plt.bar(height = frequencies, x = words)
```

### Here's a look at just the first 100 words (in case the above plot is not clickable):

```python
plt.figure(figsize =(30,10))
plt.xticks(rotation = 90)
plt.title("100 Random Word Frequencies in Romeo and Juliet", fontsize=20)
plt.ylabel("Word Frequency", fontsize=20)
plt.bar(height = frequencies[:100], x = words[:100])
```

### Most common words in Romeo and Juliet

```python
word_count = Counter(raw_text)
top_ten = word_count.most_common(10)
print(top_ten)
```

### We can see the top ten words and their counts in the list. Let's plot these on a bar graph to get a better look!

```python
plt.figure(figsize =(20,10))
plt.title('Frequencies of Top Ten Words in Romeo and Juliet', fontsize=20)
plt.xticks(rotation = 90)
plt.ylabel("Word Frequency", fontsize=20)
plt.bar(height = [top_ten[i][1] for i in range(len(top_ten))], x = [top_ten[j][0] for j in range(len(top_ten))])
```

### Unsurprisingly, the ten most common words are either articles or conjunctions. There are far more than just ten words with only a single instance in Romeo and Juliet so choosing the ten least common words would be arbitrary. These least common words are things like adjectives and nouns which are not too frequently used in speech.
